```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
preds = '../data/prediction/preprocessed/val/'
prepro = '../data/prediction/predicted/2301_1/grayscale'
gt = np.load('../data/training/test_masks_n_2.npy')
```

```python
fig, ax = plt.subplots(10, 3, figsize=(10, 30))

# label the columns
ax[0,2].set_title('Predicted')
ax[0,1].set_title('Ground Truth')
ax[0,0].set_title('Input')

idx = [2,4,9,1,0,7,8,6,5,3]

for i, num in enumerate(idx):
    ax[i,0].imshow(cv2.imread(os.path.join(preds, str(num) + '.png')))
    ax[i,2].imshow(cv2.imread(os.path.join(prepro, str(num) + '.png')))
    # colormap gray
    ax[i,1].imshow(gt[num], cmap='gray')

plt.savefig('results.png')
```
